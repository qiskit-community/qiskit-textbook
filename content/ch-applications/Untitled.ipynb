```python
from qiskit import QuantumCircuit, Aer
from qiskit.visualization import plot_state_qsphere
from math import pi

sim = Aer.get_backend('aer_simulator')

qc = QuantumCircuit(3)
qc.x(0)
qc.x(1)
qc.h(2)
theta = pi/2
qc.cp(theta,1,2)
#qc.cx(0,1)
#qc.cp(-theta,1,2)
#qc.cx(0,1)
#qc.cp(theta,0,2)
display(qc.draw())

qc.save_statevector()
result = sim.run(qc).result().get_statevector()
plot_state_qsphere(result)
```
