## 1. åˆ¶å¾¡Zã‚’CNOTã‹ã‚‰ä½œæˆã™ã‚‹ <a id="c-from-cnot"></a>

åˆ¶å¾¡Zã¾ãŸã¯`cz`ã‚²ãƒ¼ãƒˆã¯ã€ã‚ˆãä½¿ç”¨ã•ã‚Œã‚‹ã¾ãŸåˆ¥ã®2é‡å­ãƒ“ãƒƒãƒˆã‚²ãƒ¼ãƒˆã§ã™ã€‚ CNOTã¯ã€ãã®åˆ¶å¾¡é‡å­ãƒ“ãƒƒãƒˆãŒ $ |1\rangle $ã®çŠ¶æ…‹ã«ã‚ã‚‹ã¨ãã¯ãã®æ¨™çš„é‡å­ãƒ“ãƒƒãƒˆã«$X$ã‚’é©ç”¨ã—ã¾ã™ãŒã€åŒã˜ã‚ˆã†ã«ã€åˆ¶å¾¡$Z$ã¯ã€åŒã˜å ´åˆã«ğ‘ã‚’é©ç”¨ã—ã¾ã™ã€‚ Qiskitã§ã¯ã€åˆ¶å¾¡$Z$ã‚’ç›´æ¥å‘¼ã³å‡ºã™ã“ã¨ãŒã§ãã¾ã™ï¼š

```python
# åˆ¶å¾¡Z
qc.cz(c,t)
qc.draw()
```

ã“ã“ã§ã€cã¨tã¯åˆ¶å¾¡é‡å­ãƒ“ãƒƒãƒˆã¨æ¨™çš„é‡å­ãƒ“ãƒƒãƒˆã§ã™ã€‚ãŸã ã—ã€IBM Qãƒ‡ãƒã‚¤ã‚¹ã§ã¯ã€ç›´æ¥é©ç”¨ã§ãã‚‹2é‡å­ãƒ“ãƒƒãƒˆã‚²ãƒ¼ãƒˆã¯CNOTã ã‘ã§ã™ã€‚ã—ãŸãŒã£ã¦ã€å¤‰æ›ã™ã‚‹æ–¹æ³•ãŒå¿…è¦ã§ã™ã€‚

ã“ã®ãƒ—ãƒ­ã‚»ã‚¹ã¯éå¸¸ã«ç°¡å˜ã§ã™ã€‚ã‚¢ãƒ€ãƒãƒ¼ãƒ«ãŒ$|0\rangle$ãŠã‚ˆã³$|1\rangle$ã®çŠ¶æ…‹ã‚’$|+\rangle$ãŠã‚ˆã³$|-\rangle$ã®çŠ¶æ…‹ã«å¤‰æ›ã™ã‚‹ã“ã¨ã¯ã‚ã‹ã£ã¦ã„ã¾ã™ã€‚ $|+\rangle$ã¨$|-\rangle$ã®çŠ¶æ…‹ã«å¯¾ã™ã‚‹ğ‘ã‚²ãƒ¼ãƒˆã®å½±éŸ¿ã¯ã€$|0\rangle$ã¨$|1\rangle$ã®çŠ¶æ…‹ã«å¯¾ã™ã‚‹$X$ã®ä½œç”¨ã¨åŒã˜ã§ã‚ã‚‹ã“ã¨ã‚‚ã‚ã‹ã£ã¦ã„ã¾ã™ã€‚ã“ã®ã“ã¨ã‹ã‚‰ã€ã¾ãŸã¯å˜ç´”ã«è¡Œåˆ—ã‚’ä¹—ç®—ã™ã‚‹ã“ã¨ã‹ã‚‰ã€

$$
H X H = Z,\\\\
H Z H = X.
$$

åŒã˜ãƒˆãƒªãƒƒã‚¯ã‚’ä½¿ç”¨ã—ã¦ã€CNOTã‚’åˆ¶å¾¡$Z$ã«å¤‰æ›ã§ãã¾ã™ã€‚ç§ãŸã¡ãŒã—ãªã‘ã‚Œã°ãªã‚‰ãªã„ã®ã¯ã€CNOTã®å‰ã¨å¾Œã§ã€ã‚¢ãƒ€ãƒãƒ¼ãƒ«ã‚’æ¨™çš„é‡å­ãƒ“ãƒƒãƒˆã«ä½œç”¨ã•ã›ã‚‹ã ã‘ã§ã™ã€‚ã“ã‚Œã«ã‚ˆã‚Šã€ãã®é‡å­ãƒ“ãƒƒãƒˆã«é©ç”¨ã•ã‚ŒãŸ$X$ãŒ$Z$ã«å¤‰æ›ã•ã‚Œã¾ã™ã€‚

```python
qc = QuantumCircuit(2)
# ã“ã¡ã‚‰ã‚‚åˆ¶å¾¡Z
qc.h(t)
qc.cx(c,t)
qc.h(t)
qc.draw()
```

ã‚ˆã‚Šä¸€èˆ¬çš„ã«ã¯ã€å˜ä¸€ã®CNOTã®å‰å¾Œã«ã€æ­£ã—ã„å›è»¢ã‚’é…ç½®ã™ã‚‹ã“ã¨ã«ã‚ˆã‚Šã€ãƒ–ãƒ­ãƒƒãƒ›çƒã§ã®è§’åº¦$\pi$ã®ä»»æ„ã®å›è»¢ã®åˆ¶å¾¡ãƒãƒ¼ã‚¸ãƒ§ãƒ³ã«å¤‰æ›ã§ãã¾ã™ã€‚ãŸã¨ãˆã°ã€åˆ¶å¾¡$Y$ã¯ï¼š

```python
qc = QuantumCircuit(2)
# åˆ¶å¾¡Y
qc.sdg(t)
qc.cx(c,t)
qc.s(t)
qc.draw()
```

```python
qc = QuantumCircuit(2)
# åˆ¶å¾¡H
qc.ry(pi/4,t)
qc.cx(c,t)
qc.ry(-pi/4,t)
qc.draw()
```

## 2. é‡å­ãƒ“ãƒƒãƒˆã®ã‚¹ãƒ¯ãƒƒãƒ— <a id="swapping"></a>

```python
a = 0
b = 1
```

é‡å­ã‚³ãƒ³ãƒ”ãƒ¥ãƒ¼ã‚¿ãƒ¼ä¸Šã§æƒ…å ±ã‚’ç§»å‹•ã™ã‚‹å¿…è¦ãªã¨ããŒã‚ã‚Šã¾ã™ã€‚ã“ã‚Œã‚’ç‰©ç†çš„ã«ç§»å‹•ã™ã‚‹ã“ã¨ã§å®Ÿè¡Œã§ãã‚‹ã‚ˆã†ã«ã€å®Ÿè£…ã•ã‚ŒãŸé‡å­ãƒ“ãƒƒãƒˆã‚‚ã‚ã‚Šã¾ã™ãŒã€åˆ¥ã®ã‚ªãƒ—ã‚·ãƒ§ãƒ³ã¨ã—ã¦ã¯ã€2ã¤ã®é‡å­ãƒ“ãƒƒãƒˆé–“ã§çŠ¶æ…‹ã‚’ç§»å‹•ã™ã‚‹ã“ã¨ã§æƒ…å ±ã®ç§»å‹•ã‚’å®Ÿç¾ã—ã¾ã™ã€‚ã“ã‚Œã¯SWAPã‚²ãƒ¼ãƒˆã«ã‚ˆã£ã¦è¡Œã‚ã‚Œã¾ã™ã€‚

```python
qc = QuantumCircuit(2)
# é‡å­ãƒ“ãƒƒãƒˆaã¨bã®çŠ¶æ…‹ã‚’ã‚¹ãƒ¯ãƒƒãƒ—ã™ã‚‹
qc.swap(a,b)
qc.draw()
```

ä¸Šè¨˜ã®ã‚³ãƒãƒ³ãƒ‰ã¯ã“ã®ã‚²ãƒ¼ãƒˆã‚’ç›´æ¥å‘¼ã³å‡ºã—ã¾ã™ãŒã€æ¨™æº–ã®ã‚²ãƒ¼ãƒˆã‚»ãƒƒãƒˆã‚’ä½¿ç”¨ã—ã¦ã“ã‚Œã‚’ä½œæˆã™ã‚‹æ–¹æ³•ã‚’è¦‹ã¦ã¿ã¾ã—ã‚‡ã†ã€‚ã“ã®ãŸã‚ã«ã¯ã€ã„ãã¤ã‹ã®ä¾‹ã‚’æ¤œè¨ã™ã‚‹å¿…è¦ãŒã‚ã‚Šã¾ã™ã€‚

ã¾ãšã€é‡å­ãƒ“ãƒƒãƒˆaãŒ$|1\rangle$ã®çŠ¶æ…‹ã§ã€é‡å­ãƒ“ãƒƒãƒˆbãŒ$|0\rangle$ã®çŠ¶æ…‹ã§ã‚ã‚‹å ´åˆã‚’è¦‹ã¦ã¿ã¾ã—ã‚‡ã†ã€‚ã“ã‚Œã«ã¯ã€æ¬¡ã®ã‚²ãƒ¼ãƒˆã‚’é©ç”¨ã—ã¾ã™ï¼š

```python
qc = QuantumCircuit(2)
# aã®1ã‚’aã‹ã‚‰bã«äº¤æ›
qc.cx(a,b) # aã‹ã‚‰bã¸1ã‚’ã‚³ãƒ”ãƒ¼ã™ã‚‹
qc.cx(b,a) # bã®1ã‚’ä½¿ç”¨ã—ã¦aã®çŠ¶æ…‹ã‚’0ã«å›è»¢
qc.draw()
```

ã“ã‚Œã¯ã€é‡å­ãƒ“ãƒƒãƒˆbã‚’çŠ¶æ…‹$|1\rangle$ã«ã—ã€é‡å­ãƒ“ãƒƒãƒˆaã‚’çŠ¶æ…‹$|0\rangle$ã«ã™ã‚‹åŠ¹æœãŒã‚ã‚Šã¾ã™ã€‚ã“ã®å ´åˆã€å°‘ãªãã¨ã‚‚SWAPã‚’å®Ÿè¡Œã—ãŸã“ã¨ã«ãªã‚Šã¾ã™ã€‚

æ¬¡ã«ã€ã“ã®çŠ¶æ…‹ã‚’å…ƒã®çŠ¶æ…‹ã«æˆ»ã—ã¾ã™ã€‚ã”æƒ³åƒã®ã¨ãŠã‚Šã€ä¸Šè¨˜ã®ãƒ—ãƒ­ã‚»ã‚¹ã®é€†ã§ã“ã‚Œã‚’è¡Œã†ã“ã¨ãŒã§ãã¾ã™ï¼š

```python
# qã‚’bã‹ã‚‰aã«å…¥ã‚Œæ›¿ãˆã‚‹
qc.cx(b,a) # 1ã‚’bã‹ã‚‰aã«ã‚³ãƒ”ãƒ¼ã™ã‚‹
qc.cx(a,b) # aã®1ã‚’ä½¿ç”¨ã—ã¦bã®çŠ¶æ…‹ã‚’0ã«å›è»¢
qc.draw()
```

ã“ã‚Œã‚‰ã®2ã¤ã®ãƒ—ãƒ­ã‚»ã‚¹ã§ã¯ã€ä¸€æ–¹ã®æœ€åˆã®ã‚²ãƒ¼ãƒˆãŒä»–æ–¹ã®åˆæœŸçŠ¶æ…‹ã«å½±éŸ¿ã‚’ä¸ãˆãªã„ã“ã¨ã«æ³¨æ„ã—ã¦ãã ã•ã„ã€‚ãŸã¨ãˆã°ã€bã®$|1\rangle$ã‚’aã«äº¤æ›ã™ã‚‹å ´åˆã€æœ€åˆã®ã‚²ãƒ¼ãƒˆã¯`cx(b,a)`ã§ã™ã€‚ä»£ã‚ã‚Šã«ã€ã“ã‚ŒãŒæœ€åˆã«bã«$|1\rangle$ãŒãªã‹ã£ãŸçŠ¶æ…‹ã«é©ç”¨ã•ã‚ŒãŸå ´åˆã€åŠ¹æœã¯ã‚ã‚Šã¾ã›ã‚“ã€‚

ã¾ãŸã€ã“ã‚Œã‚‰ã®2ã¤ã®ãƒ—ãƒ­ã‚»ã‚¹ã§ã¯ã€ä¸€æ–¹ã®æœ€çµ‚ã‚²ãƒ¼ãƒˆãŒä»–æ–¹ã®æœ€çµ‚çŠ¶æ…‹ã«å½±éŸ¿ã‚’ä¸ãˆãªã„ã“ã¨ã«æ³¨æ„ã—ã¦ãã ã•ã„ã€‚ãŸã¨ãˆã°ã€$|1\rangle$ã‚’aã‹ã‚‰bã«ã‚¹ãƒ¯ãƒƒãƒ—ã™ã‚‹ã¨ãã«å¿…è¦ãªæœ€å¾Œã®`cx(b,a)`ã¯ã€$|1\rangle$ãŒbã«ãªã„çŠ¶æ…‹ã«ã¯å½±éŸ¿ã—ã¾ã›ã‚“ã€‚

ã“ã‚Œã‚‰ã®è¦³å¯Ÿçµæœã‚’å…ƒã«ã€ä¸€æ–¹ã®ã‚²ãƒ¼ãƒˆã‹ã‚‰ã‚‚ã†ä¸€æ–¹ã®ã‚²ãƒ¼ãƒˆã«ç„¡åŠ¹ãªã‚²ãƒ¼ãƒˆã‚’è¿½åŠ ã—ã¦ã€2ã¤ã®ãƒ—ãƒ­ã‚»ã‚¹ã‚’çµ„ã¿åˆã‚ã›ã‚‹ã“ã¨ãŒã§ãã¾ã™ã€‚ä¾‹ãˆã°ã€

```python
qc = QuantumCircuit(2)
qc.cx(b,a)
qc.cx(a,b)
qc.cx(b,a)
qc.draw()
```

ã“ã‚Œã¯ã€aã‹ã‚‰bã«$|1\rangle$ã‚’äº¤æ›ã™ã‚‹ãƒ—ãƒ­ã‚»ã‚¹ã¨è€ƒãˆã‚‹ã“ã¨ãŒã§ãã¾ã™ãŒã€æœ€åˆã¯å½¹ã«ç«‹ãŸãªã„ `qc.cx(b,a)`ãŒã‚ã‚Šã¾ã™ã€‚ã¾ãŸã€$|1\rangle$ã‚’bã‹ã‚‰aã«äº¤æ›ã™ã‚‹ãƒ—ãƒ­ã‚»ã‚¹ã¨è€ƒãˆã‚‹ã“ã¨ã‚‚ã§ãã¾ã™ãŒã€æœ€å¾Œã«ç„¡ç”¨ãª`qc.cx(b,a)`ãŒã‚ã‚Šã¾ã™ã€‚ã©ã¡ã‚‰ã«ã—ã¦ã‚‚ã€çµæœã¯ã€åŒæ–¹å‘ã§ã‚¹ãƒ¯ãƒƒãƒ—ã‚’å®Ÿè¡Œã§ãã‚‹ãƒ—ãƒ­ã‚»ã‚¹ã§ã™ã€‚

ã¾ãŸã€çŠ¶æ…‹$|00\rangle$ã«å¯¾ã—ã¦æ­£ã—ãåŠ¹æœã‚’ç™ºæ®ã—ã¾ã™ã€‚ã“ã‚Œã¯å¯¾ç§°çš„ã§ã‚ã‚‹ãŸã‚ã€çŠ¶æ…‹ã‚’å…¥ã‚Œæ›¿ãˆã¦ã‚‚å½±éŸ¿ã¯ã‚ã‚Šã¾ã›ã‚“ã€‚åˆ¶å¾¡é‡å­ãƒ“ãƒƒãƒˆãŒ$|0\rangle$ã®å ´åˆã€CNOTã‚²ãƒ¼ãƒˆã¯åŠ¹æœãŒãªã„ãŸã‚ã€ãƒ—ãƒ­ã‚»ã‚¹ã¯ä½•ã‚‚ã—ã¾ã›ã‚“ã€‚

çŠ¶æ…‹$|11\rangle$ã‚‚å¯¾ç§°çš„ã§ã‚ã‚‹ãŸã‚ã€ã‚¹ãƒ¯ãƒƒãƒ—ã«ã‚ˆã‚‹åŠ¹æœã¯ãªã„ã«ç­‰ã—ã„ã§ã™ã€‚ã“ã®å ´åˆã€ä¸Šè¨˜ã®ãƒ—ãƒ­ã‚»ã‚¹ã®æœ€åˆã®CNOTã‚²ãƒ¼ãƒˆã¯2ç•ªç›®ã®ã‚²ãƒ¼ãƒˆã«å½±éŸ¿ã‚’ä¸ãˆãšã€3ç•ªç›®ã®ã‚²ãƒ¼ãƒˆã¯æœ€åˆã®ã‚²ãƒ¼ãƒˆã‚’å…ƒã«æˆ»ã—ã¾ã™ã€‚ã—ãŸãŒã£ã¦ã€å…¨ä½“ã®åŠ¹æœã¯ç¢ºã‹ã«ãªã„ã«ç­‰ã—ã„ã§ã™ã€‚

ã“ã®ã‚ˆã†ã«ã—ã¦ã€SWAPã‚²ãƒ¼ãƒˆã‚’æ¨™æº–ã‚²ãƒ¼ãƒˆã‚»ãƒƒãƒˆã§ã‚ã‚‹å˜ä¸€é‡å­ãƒ“ãƒƒãƒˆã®å›è»¢ã‚²ãƒ¼ãƒˆã¨CNOTã‚²ãƒ¼ãƒˆã«åˆ†è§£ã™ã‚‹æ–¹æ³•ã‚’è¦‹ã¤ã‘ã¾ã—ãŸã€‚

```python
qc = QuantumCircuit(2)
# é‡å­ãƒ“ãƒƒãƒˆaã¨bã®çŠ¶æ…‹ã®ã‚¹ãƒ¯ãƒƒãƒ—
qc.cx(b,a)
qc.cx(a,b)
qc.cx(b,a)
qc.draw()
```

CNOTã‚²ãƒ¼ãƒˆã®é †åºã‚’å¤‰æ›´ã—ãŸå ´åˆã‚‚ã€åŒã˜åŠ¹æœãŒå¾—ã‚‰ã‚Œã¾ã™ï¼š

```python
qc = QuantumCircuit(2)
# é‡å­ãƒ“ãƒƒãƒˆaã¨bã®çŠ¶æ…‹ã‚’å…¥ã‚Œæ›¿ãˆã¾ã™
qc.cx(a,b)
qc.cx(b,a)
qc.cx(a,b)
qc.draw()
```

## 3. åˆ¶å¾¡å›è»¢ <a id="controlled-rotations"></a>

åˆ¶å¾¡$\pi$å›è»¢ã‚’å˜ä¸€ã®CNOTã‚²ãƒ¼ãƒˆã‹ã‚‰æ§‹ç¯‰ã™ã‚‹æ–¹æ³•ã«ã¤ã„ã¦ã‚’ã¿ã¦ãã¾ã—ãŸã€‚ æ¬¡ã«ã€ä»»æ„ã®åˆ¶å¾¡å›è»¢ã‚’æ§‹ç¯‰ã™ã‚‹æ–¹æ³•ã‚’ã¿ã¦ã„ãã¾ã™ã€‚

ã¾ãšã€yè»¸ã‚’ä¸­å¿ƒã¨ã—ãŸä»»æ„ã®å›è»¢ã‚’è€ƒãˆã¦ã¿ã¾ã—ã‚‡ã†ã€‚ å…·ä½“çš„ã«ã¯ã€æ¬¡ã®ä¸€é€£ã®ã‚²ãƒ¼ãƒˆã‚’è€ƒãˆã¦ã¿ã¦ãã ã•ã„ã€‚

```python
qc = QuantumCircuit(2)
theta = pi # thetaã¯ä½•ã§ã‚‚ã‚ˆã„ (ä»»æ„ã®å€¤ã¨ã—ã¦piã‚’é¸æŠ)
qc.ry(theta/2,t)
qc.cx(c,t)
qc.ry(-theta/2,t)
qc.cx(c,t)
qc.draw()
```

ã“ã®å›è·¯ã¯ã€åˆ¶å¾¡é‡å­ãƒ“ãƒƒãƒˆãŒçŠ¶æ…‹ $|0\rangle$ã®å ´åˆã€$R_y(\theta/2)$ã®å¾Œã«ãã®é€†ã® $R_y(-\theta/2)$ ãŒç¶šãã ã‘ã§ã™ã€‚ã“ã®æœ€çµ‚çš„ãªåŠ¹æœã¯ã€ãªã„ã«ç­‰ã—ã„ã§ã™ã€‚ã—ã‹ã—ã€åˆ¶å¾¡é‡å­ãƒ“ãƒƒãƒˆãŒ $|1\rangle$ ã®å ´åˆã€`ry(-theta/2)`ã®å‰å¾Œã«Xã‚²ãƒ¼ãƒˆãŒé©ç”¨ã•ã‚Œã¾ã™ã€‚ã“ã‚Œã¯ã€yå›è»¢ã®æ–¹å‘ã‚’åè»¢ã—ã€2ç•ªç›®ã® $R_y(\theta/2)$ ã‚’ä½œæˆã™ã‚‹åŠ¹æœãŒã‚ã‚Šã¾ã™ã€‚ã—ãŸãŒã£ã¦ã€ã“ã®å ´åˆã®æ­£å‘³ã®ä½œç”¨ã¯ã€åˆ¶å¾¡ãƒãƒ¼ã‚¸ãƒ§ãƒ³ã®å›è»¢ $R_y(\theta)$ ã‚’ä½œã‚‹ã“ã¨ã¨åŒã˜ã«ãªã‚Šã¾ã™ã€‚

ã“ã®æ–¹æ³•ãŒæ©Ÿèƒ½ã™ã‚‹ã®ã¯ã€xã¨yè»¸ãŒç›´è¡Œã—ã€ãã®ãŸã‚ã«xã‚²ãƒ¼ãƒˆãŒå›è»¢ã®æ–¹æ³•ã‚’åè»¢ã•ã›ã‚‹ãŸã‚ã§ã™ã€‚ã—ãŸãŒã£ã¦ã€åˆ¶å¾¡$R_z(\theta)$ã‚’ä½œã‚‹ã®ã‚‚åŒã˜ã‚ˆã†ã«æ©Ÿèƒ½ã—ã¾ã™ã€‚åˆ¶å¾¡ $R_x(\theta)$ ã¯ã€CNOTã‚²ãƒ¼ãƒˆã‚’ä½¿ã£ã¦åŒã˜ã‚ˆã†ã«ä½œã‚‹ã“ã¨ãŒã§ãã¾ã™ã€‚

ã¾ãŸã€å˜ä¸€é‡å­ãƒ“ãƒƒãƒˆå›è»¢$U$ã®åˆ¶å¾¡ãƒãƒ¼ã‚¸ãƒ§ãƒ³ã‚‚ä½œã‚‹ã“ã¨ãŒã§ãã¾ã™ã€‚ãã®ãŸã‚ã«ã¯ã€3ã¤ã®å›è»¢Aã€Bã€Cã¨ã€æ¬¡ã®ã‚ˆã†ãªä½ç›¸ $\alpha$ ã‚’è¦‹ã¤ã‘ã‚‹ã ã‘ã§ã™ã€‚

$$
ABC = I, ~~~e^{i\alpha}AZBZC = U
$$

åˆ¶å¾¡Zã‚²ãƒ¼ãƒˆã‚’ä½¿ã†ã“ã¨ã§ã€ä¸Šã®å·¦ã®å¼ã‚’åˆ¶å¾¡ãƒ“ãƒƒãƒˆãŒ$|0\rangle$ã®çŠ¶æ…‹ã®éš›ã«ç™ºç”Ÿã•ã›ã€å³ã®å¼ã‚’åˆ¶å¾¡ãƒ“ãƒƒãƒˆãŒ$|1\rangle$ã®çŠ¶æ…‹ã®ã¨ãã«ç™ºç”Ÿã•ã›ã¾ã™ã€‚æ­£ã—ã„ä½ç›¸ã‚’å¾—ã‚‹ãŸã‚ã«åˆ¶å¾¡ãƒ“ãƒƒãƒˆå´ã§$R_z(2\alpha)$å›è»¢ã‚‚ä½¿ã‚ã‚Œã€ã“ã‚Œã¯ã€é‡ã­åˆã‚ã›çŠ¶æ…‹ãŒã‚ã‚‹å ´åˆã«é‡è¦ã§ã™ã€‚

```python
A = Gate('A', 1, [])
B = Gate('B', 1, [])
C = Gate('C', 1, [])
alpha = 1 # å›è·¯æå†™ã®ãŸã‚ã«ä»»æ„ã®alphaã‚’è¨­å®š
```

```python
qc = QuantumCircuit(2)
qc.append(C, [t])
qc.cz(c,t)
qc.append(B, [t])
qc.cz(c,t)
qc.append(A, [t])
qc.p(alpha,c)
qc.draw()
```

## 4. ãƒˆãƒ•ã‚©ãƒª <a id="ccx"></a>

ãƒˆãƒ•ã‚©ãƒªã‚²ãƒ¼ãƒˆã¯ã€2ã¤ã®åˆ¶å¾¡ã¨1ã¤ã®æ¨™çš„ã‚’æŒã¤ï¼“é‡å­ãƒ“ãƒƒãƒˆã‚²ãƒ¼ãƒˆã§ã™ã€‚ä¸¡æ–¹ã®åˆ¶å¾¡ãƒ“ãƒƒãƒˆãŒ $|1\rangle$ ã®çŠ¶æ…‹ã®æ™‚ã®ã¿æ¨™çš„ãƒ“ãƒƒãƒˆã«Xã‚’é©ç”¨ã—ã¾ã™ã€‚æ¨™çš„ãƒ“ãƒƒãƒˆã®æœ€çµ‚çŠ¶æ…‹ã¯ã€æ¨™çš„ãƒ“ãƒƒãƒˆã®åˆæœŸçŠ¶æ…‹ãŒ $|0\rangle$ ã¾ãŸã¯ $|1\rangle$ ã®ã©ã¡ã‚‰ã§ã‚ã£ãŸã‹ã«å¿œã˜ã¦ã€2ã¤ã®åˆ¶å¾¡ãƒ“ãƒƒãƒˆã®ANDã¾ãŸã¯NANDã®ã„ãšã‚Œã‹ã«ç­‰ã—ããªã‚Šã¾ã™ã€‚ãƒˆãƒ•ã‚©ãƒªã¯ã€åˆ¶å¾¡ãƒ»åˆ¶å¾¡NOTã¨ã‚‚è€ƒãˆã‚‹ã“ã¨ã‚‚ã§ãã€CCXã‚²ãƒ¼ãƒˆã¨ã‚‚å‘¼ã°ã‚Œã¾ã™ã€‚ 

```python
qc = QuantumCircuit(3)
a = 0
b = 1
t = 2
# åˆ¶å¾¡ãƒ“ãƒƒãƒˆaã€bã€æ¨™çš„ãƒ“ãƒƒãƒˆtã®ãƒˆãƒ•ã‚©ãƒª
qc.ccx(a,b,t)
qc.draw()
```

å˜ä¸€é‡å­ãƒ“ãƒƒãƒˆã‚²ãƒ¼ãƒˆã¨2é‡å­ãƒ“ãƒƒãƒˆã‚²ãƒ¼ãƒˆã‹ã‚‰ã“ã‚Œã‚’æ§‹ç¯‰ã™ã‚‹æ–¹æ³•ã‚’ç¢ºèªã™ã‚‹ãŸã‚ã«ã€ã¾ãšã€ã‚ˆã‚Šä¸€èˆ¬çš„ãªã‚‚ã®ã‚’æ§‹ç¯‰ã™ã‚‹æ–¹æ³•ã‚’ç¤ºã—ã¾ã™ï¼šä»»æ„ã®å˜ä¸€é‡å­ãƒ“ãƒƒãƒˆå›è»¢Uã«ã¤ã„ã¦ã®ã€ä»»æ„ã®åˆ¶å¾¡ãƒ»åˆ¶å¾¡Uã§ã™ã€‚ãã®ãŸã‚ã«ã¯ã€åˆ¶å¾¡ãƒãƒ¼ã‚¸ãƒ§ãƒ³ã® $V = \sqrt{U}$ ã¨ $V^\dagger$ ã®å®šç¾©ãŒå¿…è¦ã§ã™ã€‚ä»¥ä¸‹ã®ã‚³ãƒ¼ãƒ‰ã§ã¯ã€æœªå®šç¾©ã®ã‚µãƒ–ãƒ«ãƒ¼ãƒãƒ³ `cv` ã¨ `cvdg` ã®ä»£ã‚ã‚Šã«ã€ãã‚Œãã‚Œ `cu1(theta,c,t)` ã¨ `cu1(-theta,c,t)` ã‚’ä½¿ã£ã¦ã„ã¾ã™ã€‚åˆ¶å¾¡ãƒ“ãƒƒãƒˆã¯é‡å­ãƒ“ãƒƒãƒˆ $a$ ã¨ $b$ã§ã€æ¨™çš„ãƒ“ãƒƒãƒˆã¯é‡å­ãƒ“ãƒƒãƒˆ $t$ ã§ã™ã€‚

```python
qc = QuantumCircuit(3)
qc.cp(theta,b,t)
qc.cx(a,b)
qc.cp(-theta,b,t)
qc.cx(a,b)
qc.cp(theta,a,t)
qc.draw()
```

ä¾‹ãˆã°ã€åˆ¶å¾¡ã‚¢ãƒ€ãƒãƒ¼ãƒ«ã‚²ãƒ¼ãƒˆã¨åˆ¶å¾¡$Z$ã‚²ãƒ¼ãƒˆã‚’ä½¿ã†ã¨ã—ã¾ã™ã€‚ã“ã‚Œã‚‰ã®ã‚²ãƒ¼ãƒˆã¯ã€ä¸¡æ–¹ã¨ã‚‚1ã¤ã®CNOTã§å®Ÿè£…ã§ãã¾ã™ã€‚ã“ã‚Œã‚‰ã®ã‚²ãƒ¼ãƒˆã‹ã‚‰æ¬¡ã®å›è·¯ã‚’ä½œæˆã§ãã¾ã™ï¼š

```python
qc = QuantumCircuit(3)
qc.ch(a,t)
qc.cz(b,t)
qc.ch(a,t)
qc.draw()
```

## 5. Hã¨Tã«ã‚ˆã‚‹ä»»æ„ã®å›è»¢ <a id="arbitrary-rotations"></a>

ç¾åœ¨ã®ãƒ‡ãƒã‚¤ã‚¹ã®é‡å­ãƒ“ãƒƒãƒˆã¯ãƒã‚¤ã‚ºã®å½±éŸ¿ã‚’å—ã‘ã‚„ã™ãã€åŸºæœ¬çš„ã«ã¯ã‚¨ãƒ©ãƒ¼ã‚’èµ·ã“ã™ã‚²ãƒ¼ãƒˆã§æ§‹æˆã•ã‚Œã¦ã„ã¾ã™ã€‚æ¸©åº¦ã€ç£å ´ã®ã‚†ã‚‰ãã€éš£ã®é‡å­ãƒ“ãƒƒãƒˆã®å‹•ããªã©ã®ã‚ˆã†ãªå˜ç´”ãªã“ã¨ãŒã€æ„å›³ã—ã¦ã„ãªã„ã“ã¨ã‚’å¼•ãèµ·ã“ã™å¯èƒ½æ€§ãŒã‚ã‚Šã¾ã™ã€‚

é‡å­ã‚³ãƒ³ãƒ”ãƒ¥ãƒ¼ã‚¿ãƒ¼ã®å¤§è¦æ¨¡ãªã‚¢ãƒ—ãƒªã‚±ãƒ¼ã‚·ãƒ§ãƒ³ã§ã¯ã€ã“ã®ãƒã‚¤ã‚ºã‹ã‚‰é‡å­ãƒ“ãƒƒãƒˆã‚’ä¿è­·ã™ã‚‹ã‚ˆã†ã«é‡å­ãƒ“ãƒƒãƒˆã‚’ã‚¨ãƒ³ã‚³ãƒ¼ãƒ‰ã™ã‚‹å¿…è¦ãŒã‚ã‚Šã¾ã™ã€‚ã“ã‚Œã¯ã€ã‚²ãƒ¼ãƒˆã‚’é–“é•ã£ã¦å®Ÿè¡Œã™ã‚‹ã“ã¨ã‚’é›£ã—ãã™ã‚‹ã‹ã€ã¾ãŸã¯ã‚ãšã‹ã«é–“é•ã£ãŸæ–¹æ³•ã§å®Ÿè£…ã™ã‚‹ã“ã¨ã«ã‚ˆã£ã¦è¡Œã‚ã‚Œã¾ã™ã€‚

å˜ä¸€é‡å­ãƒ“ãƒƒãƒˆã®å›è»¢$R_x(\theta)$ã€ $R_y(\theta)$ ã€ $R_z(\theta)$ ã¯æ®‹å¿µã§ã™ãŒã€èª¤ã£ã¦$\theta + 0.0000001$ã®ã‚ˆã†ãªå®Ÿè£…ã‚’ã—ã¦ã„ãªã„ã“ã¨ã‚’ç¢ºä¿¡ã§ãã‚‹ã‚ˆã†ãªå®Œç’§ãªç²¾åº¦ã§è§’åº¦$\theta$ã‚’ã«å®Ÿè£…ã™ã‚‹ã“ã¨ã¯ä¸å¯èƒ½ã§ã™ã€‚é”æˆã§ãã‚‹ç²¾åº¦ã«ã¯å¸¸ã«é™ç•ŒãŒã‚ã‚Šã€å¤§ããªå›è·¯ã§ã¯ã‚¨ãƒ©ãƒ¼ã®ç©ã¿é‡ã­ãŒã‚ã‚‹ã“ã¨ã‚’è€ƒæ…®ã™ã‚‹ã¨ã€ãã®ç²¾åº¦ã¯è¨±å®¹ç¯„å›²ã‚’å¸¸ã«è¶…ãˆã¦ã„ã¾ã™ã€‚ã—ãŸãŒã£ã¦ã€ã“ã‚Œã‚‰ã®å›è»¢ã‚’ãƒ•ã‚©ãƒ¼ãƒ«ãƒˆãƒˆãƒ¬ãƒ©ãƒ³ãƒˆãªé‡å­ã‚³ãƒ³ãƒ”ãƒ¥ãƒ¼ã‚¿ãƒ¼ã«ç›´æ¥å®Ÿè£…ã™ã‚‹ã“ã¨ã¯ã§ãã¾ã›ã‚“ãŒã€ä»£ã‚ã‚Šã«ã€ã‚ˆã‚Šå·¥å¤«ã—ãŸæ–¹æ³•ã§å›è»¢ã‚’æ§‹ç¯‰ã™ã‚‹å¿…è¦ãŒã‚ã‚Šã¾ã™ã€‚

ãƒ•ã‚©ãƒ¼ãƒ«ãƒˆãƒˆãƒ¬ãƒ©ãƒ³ãƒˆã®ã‚¹ã‚­ãƒ¼ãƒ ã§ã¯ã€é€šå¸¸ã€2ã¤ã®ã¿ã®ã‚²ãƒ¼ãƒˆã€$H$ ã¨ $T$ ã‚’è¤‡æ•°é©ç”¨ã™ã‚‹ã“ã¨ã«ã‚ˆã£ã¦ã“ã‚Œã‚‰ã®å›è»¢ã‚’å®Ÿè¡Œã—ã¾ã™ã€‚

Tã‚²ãƒ¼ãƒˆã¯ã€Qiskitã§ã¯`.t()`ã¨ã—ã¦è¡¨ç¾ã•ã‚Œã¾ã™ï¼š

```python
qc = QuantumCircuit(1)
qc.t(0) # é‡å­ãƒ“ãƒƒãƒˆ0ã¸Tã‚²ãƒ¼ãƒˆã‚’é©ç”¨
qc.draw()
```

ã“ã‚Œã¯zè»¸ã‚’ä¸­å¿ƒã¨ã—ãŸ $\theta = \pi/4$ ã®å›è»¢ãªã®ã§ã€æ•°å­¦çš„ã«ã¯ $R_z(\pi/4) = e^{i\pi/8~Z}$ ã¨è¡¨ã•ã‚Œã¾ã™ã€‚

ä»¥ä¸‹ã§ã¯ã€$H$ ã¨ $T$ ã‚²ãƒ¼ãƒˆã®åŠ¹æœãŒå®Œå…¨ã§ã‚ã‚‹ã¨ä»®å®šã—ã¾ã™ã€‚ãã‚Œã¯ã€ã‚¨ãƒ©ãƒ¼è¨‚æ­£ã¨ãƒ•ã‚©ãƒ¼ãƒ«ãƒˆãƒˆãƒ¬ãƒ©ãƒ³ã‚¹ã«é©ã—ãŸæ–¹æ³•ã§è¨­è¨ˆã§ãã¾ã™ã€‚

ã‚¢ãƒ€ãƒãƒ¼ãƒ«ã¨å‰ã®ç« ã§èª¬æ˜ã—ãŸæ–¹æ³•ã‚’ä½¿ç”¨ã™ã‚‹ã¨ã€Tã‚²ãƒ¼ãƒˆã‚’ä½¿ã£ã¦ã€xè»¸ã‚’ä¸­å¿ƒã«åŒæ§˜ã®å›è»¢ã‚’ä½œæˆã§ãã¾ã™ã€‚

```python
qc = QuantumCircuit(1)
qc.h(0)
qc.t(0)
qc.h(0)
qc.draw()
```
